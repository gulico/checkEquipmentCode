# 全厂码检查

筛选查看“全厂码”列，下拉选项中只有G01和/或T01，多台风机有多个全厂码，G**或G***，注意序列号要与风机编号对应。

```python
groupby_QCM = sheet1.groupby('全厂码F0')
```

```python
for name, group in groupby_QCM:
    if not (name[0] == 'T' or name[0] == 'G'):
        print('错误全厂码F0：' + name)
        print('全厂码F0首字母不为G或T，请检查全厂码F0首字母')
    if not (name[1:].isdigit()):
        print('错误全厂码F0：' + name)
        print('全厂码F0除第一位外不为数字，请调整为数字')
    if not len(name) == 3:
        print('错误全厂码F0：' + name)
        print('全厂码F0长度不为3，请调整全厂码F0长度为3')
```

```python
print(groupby_QCM.size())
print('全厂码个数：' + str(groupby_QCM.ngroups))
if not (groupby_QCM.ngroups) == 1:  #分组个数只为1
    print('每个sheet全厂码F0应只有1种，不同全场码F0请分成不同sheet')
```

# 系统码检查

- 筛选查看“系统码”列，下拉选项中的系统码全部为3个字母+2位数字，
- 所用系统码均在树状图模板中有。

## 系统码全部为3个字母+2位数字

```python
groupby_XTM = sheet1.groupby('系统码F1')
```

```python
for name, group in groupby_XTM:
    if not (name[0:3].isalpha()):
        print('错误系统码F1：' + name)
        print('系统码F1前3位应全为字母')
    if not (name[3:5].isdigit()):
        print('错误系统码F1：' + name)
        print('系统码F1后2位应全为数字')
    if not len(name) == 5:
        print('错误系统码F1：' + name)
        print('系统码F1长度不为5，请调整系统码F1长度为5')
```

## 系统码均在树状图模板中有

```python
# 选择系统码树状图模板
tree_tmp = []
if DC_type == '风电':
    tree_tmp = (XTM_tree_FD['系统码']).to_list()
elif DC_type == '光伏':
    tree_tmp = (XTM_tree_GF['系统码']).to_list()

# 检查系统码是否在树状图中
for index, row in sheet1.iterrows():
    if not(isinstance(row['系统码F1'],str)):  #跳过NAN空值
        continue
    if not (row['系统码F1'] in tree_tmp):
        print('系统码F1在树状图模板中不存在！ 序号：', row['序号'], '系统码F1：', row['系统码F1'])
```

# 设备码检查

- 筛选查看“设备码”列，下拉选项中的设备码全部为2个字母+3位数字，
- 所用设备码均在设备码模板中有，
- 编号都从001开始。
- 光伏逆变器、汇流箱、组串和支架等有特殊规则，按其规则仔细检查。

## 风电设备码全部为2个字母+3位数字

```python
groupby_SBM = sheet1.groupby('设备码F2')
```

```python
if DC_type == '风电':
    for name, group in groupby_SBM:
        if not (name[0:2].isalpha()):
            print('错误设备码F2：' + name)
            print('设备码F2前2位应全为字母')
        if not (name[2:5].isdigit()):
            print('错误设备码F2：' + name)
            print('设备码F2后3位应全为数字')
        if not len(name) == 5:
            print('错误设备码F2：' + name)
            print('设备码F2长度不为5，请调整设备码F2长度为5')
```

## 所用设备码均在设备码模板中有

```python
tree_tmp = SBM_tree['设备/产品分类码'].to_list()

# 检查设备码是否在树状图中
for index, row in sheet1.iterrows():
    if not(isinstance(row['设备码F2'],str)):  #跳过NAN空值
        continue
    if not (row['设备码F2'][0:2] in tree_tmp):
        print('设备码F2在设备编码模板中不存在！ 序号：', row['序号'], '设备码F2：', row['设备码F2'])
```

## 光伏逆变器、汇流箱、组串和支架等有特殊规则，按其规则仔细检查。

```python
SBM_list = []
for name, group in groupby_SBM:
    SBM_list.append(name)
```

```python
if DC_type == '光伏':
    GC_bit4 = False
    GC_bit6 = False
    for index, row in sheet1.iterrows():
        if not(isinstance(row['设备码F2'],str)):  #跳过NAN空值
            continue
            
        name = row['设备码F2']
        numb = row['序号']
####################### 光伏组件 #######################
        if name[0:2] == 'GC':  # 光伏组件
            if len(name[2:])==4:  # 光伏组件4位设备码
                GC_bit4 = True
                if not ('TB'+str(name[2:4]) in SBM_list):
                    print('序号：',numb,'光伏组件设备码',name,'缺少对应逆变器','TB'+str(name[2:4]))
            elif len(name[2:])==6: # 光伏组件6位设备码
                GC_bit6 = True
                TB4 = 'TB'+str(name[2:6]) in SBM_list
                TB2 = 'TB'+str(name[2:4]) in SBM_list
                UC4 = 'UC'+str(name[2:6]) in SBM_list
                UC2 = 'UC'+str(name[2:4]) in SBM_list

                if TB4: # 有逆变器4位
                    if not UC2: # 有逆变器4位，无汇流箱2位
                        print('序号：',numb,'光伏组件设备码',name,'缺少对应汇流箱','UC'+str(name[2:4]))
                else: # 没有4位逆变器
                    if UC4: # 汇流箱4位
                        if not TB2: # 有汇流箱4位，无逆变器2位
                            print('序号：',numb,'光伏组件设备码',name,'缺少对应逆变器','TB'+str(name[2:4]))
                    else: # 没有4位逆变器，没有4位汇流箱
                        if not (UC2 or TB2): # 没有2位逆变器，没有2位汇流箱
                            print('序号：',numb,'光伏组件设备码',name,'缺少对应逆变器和汇流箱')
                        elif UC2:
                            print('序号：',numb,'光伏组件设备码',name,'有汇流箱','UC'+str(name[2:4]),'缺少对应4位逆变器')
                        elif TB2:
                            print('序号：',numb,'光伏组件设备码',name,'有逆变器','TB'+str(name[2:4]),'缺少对应4位汇流箱')
            else:  #光伏组件长度错误
                print('错误设备码F2：' + name)
                print('序号：',numb,'光伏组件设备码F2数字部分应该为4位或6位')
####################### 逆变器 #######################
        elif name[0:2] == 'TB'and '逆变器' in row['设备名称']:  #逆变器
            if len(name[2:])==4:
                UC2 = 'UC'+str(name[2:4]) in SBM_list
                if not UC2: # 有逆变器4位，无汇流箱2位
                    print('序号：',numb,'光伏组件设备码',name,'缺少对应汇流箱','UC'+str(name[2:4]))
            elif not (len(name[2:])==2):  #逆变器长度错误
                print('序号：',numb,'错误设备码F2：' ,name,'逆变器设备码F2数字部分应该为2位或4位')
####################### 汇流箱 #######################
        elif name[0:2] == 'UC' and '汇流箱' in row['设备名称']:  #汇流箱
            if len(name[2:])==4:
                TB2 = 'TB'+str(name[2:4]) in SBM_list
                if not TB2: # 有汇流箱4位，无逆变器2位
                    print('序号：',numb,'光伏组件设备码',name,'缺少对应汇流箱','TB'+str(name[2:4]))
            elif not (len(name[2:])==2):  #汇流箱长度错误
                print('序号：',numb,'错误设备码F2：' , name,'汇流箱设备码F2数字部分应该为2位或4位')
####################### 其他 #######################
        elif not name[0:2] == 'UR':
            if not (name[0:2].isalpha()):
                print('序号：',numb,'错误设备码F2：' , name,'设备码F2前2位应全为字母')
            if not (name[2:5].isdigit()):
                print('序号：',numb,'错误设备码F2：' , name,'设备码F2后3位应全为数字')
            if not len(name) == 5:
                print('序号：',numb,'错误设备码F2：' , name,'设备码F2长度不为5，请调整设备码F2长度为5')
####################### 支架 #######################              
    for index, row in sheet1.iterrows():
        if name[0:2] == 'UR':  #支架
            if not ((GC_bit4 and len(name[2:])==4) or (GC_bit6 and len(name[2:])==6)):
                print('序号：',numb,'错误设备码F2：' , name,'支架编码长度必须与光伏组件统一')
```

# 产品码检查

- 筛选查看“产品码”第1列，下拉选项中的产品码全部为2个字母+2位数字，所用设备码均在设备码模板中有，编号都从01开始；
- “产品码”第2列也做相同检查。
- 5级设备超过100个时，产品码序列号可以有3位数字。

## 前两位

```python
tree_tmp = SBM_tree['设备/产品分类码'].to_list()

# 检查设备码是否在编码模板中
for index, row in sheet1.iterrows():
    if not(isinstance(row['产品码P1'],str)):  #跳过NAN空值
        continue
    if not (row['产品码P1'][0:2] in tree_tmp):
        print('产品码P1在设备编码模板中不存在！ 序号：', row['序号'], '产品码P1：', row['产品码P1'])
        
    if not(isinstance(row['产品码P2'],str)):  #跳过NAN空值
        continue
    if not (row['产品码P2'][0:2] in tree_tmp):
        print('产品码P2在设备编码模板中不存在！ 序号：', row['序号'], '产品码P1：', row['产品码P2'])
```

## 后2-3位

```python
tmpsheet = sheet1.fillna('-')
for index, row in tmpsheet.iterrows():
    numb = row['序号']
####################### 产品码P1 #######################
    if row['产品码P1']=='-':  #跳过NAN空值
        continue
    if len(row['产品码P1'][2:])==2:
        continue
    elif len(row['产品码P1'][2:])==3:
        cpm2 = row['产品码P1'][0:2] # 前两位
        contain_cpm2_len = len(tmpsheet[tmpsheet['产品码P1'].str.contains(cpm2)])
        if contain_cpm2_len < 100:
            print('序号：',numb,'错误产品码P1：' , row['产品码P1'],'5级设备超过100个时，产品码序列号可以有3位数字,否则应为2位数字')
    else:
        print('序号：',numb,'错误产品码P1：' , row['产品码P1'],'产品码P1应为2个字母+2位数字或2个字母+3位数字')
 ####################### 产品码P2 #######################
    if row['产品码P2']=='-':  #过NAN空值
        continue
    if len(row['产品码P2'][2:])==2:
        continue
    elif len(row['产品码P2'][2:])==3:
        cpm2 = row['产品码P2'][0:2] # 前两位
        contain_cpm2_len = len(tmpsheet[tmpsheet['产品码P2'].str.contains(cpm2)])
        if contain_cpm2_len < 100:
            print('序号：',numb,'错误产品码P2：' , row['产品码P2'],'5级设备超过100个时，产品码序列号可以有3位数字,否则应为2位数字')
    else:
        print('序号：',numb,'错误产品码P2：' , row['产品码P2'],'产品码P2应为2个字母+2位数字或2个字母+3位数字')
```

# 设备层级检查

- 筛选查看“设备层级”列，下拉选项中有且只有1~5级
- 检查筛选出的编码组成部分（工厂码、全厂码、系统码、设备码、产品码有无内容）与所勾选的设备层级是否匹配，如勾选第4级，则设备码列不能为空，产品码内不得有内容
- 同时检查上级设备编码与所勾选的设备层级是否匹配，上级设备码应该比所勾选的层级少1级

## 筛选查看“设备层级”列，下拉选项中有且只有1~5级

```python
groupby_SBCJ = sheet1.groupby('设备层级')
Sum_CJ = 0
for name, group in groupby_SBCJ:
    Sum_CJ = Sum_CJ + name
    if not (name in [1,2,3,4,5]):
        print('错误的层级',name,'设备层级有且只有1~5级')
if not (Sum_CJ==15):
    print('缺少层级，设备层级有且只有1~5级')
```

## 检查筛选出的编码组成部分与所勾选的设备层级是否匹配

```python
CJM_name_list = ['工厂码U1', '全厂码F0', '系统码F1', '设备码F2', '产品码P1']
for index, row in sheet1.iterrows():
    sbcj = row['设备层级']  # 设备层级1,2,3,4,5
    numb = row['序号']
    for cjEle in CJM_name_list[0:sbcj]:  # 遍历应该有内容的层级
        if (row[cjEle + ''] != row[cjEle]):  #为nan
            print('序号：', numb, '缺少设备层级：', cjEle, '设备码组合：',row['组合'])

    for cjEle in CJM_name_list[sbcj:]:  # 遍历不应该有内容的层级
        if not (row[cjEle] != row[cjEle]):  #不为nan
            print('序号：', numb, '多余的设备层级：', cjEle, '设备码组合：',row['组合'])  
```
